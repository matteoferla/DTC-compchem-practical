In a previous notebook, there was something something, 'wet lab chemists want to use Excel spreadsheets'. Whelp, now we have to deal with that: the Moonshot data is exported as an Excel spreadsheet.
To counter that I have taken the original Excel and split it up as fake database tables.

The data was shared first in https://github.com/postera-ai/COVID_moonshot_submissions,
then I cleaned it in https://github.com/matteoferla/COVID-Moonshot-metadata-analysis
to make some analysis for the [Moonshot paper](https://www.science.org/doi/10.1126/science.abo7201),
and now I have exported it as a CSV for this practical.

```python
import pandas as pd

moonshot = pd.read_pickle('moonshot_submissions.pkl.gz')
moonshot = moonshot.loc[moonshot.okay]
moonshot = moonshot.rename(columns={'CID (canonical)': 'Id'}).reset_index(drop=True).copy()

moonshot[['Id',              # CID
          'SMILES',          # SMILES as submitted
          'IC50',            # Rapidfire MS assay (cleavage assay, not affinity-MS)
          'Structure ID',    # Internal Crystal structure Id (XChem code)
          'pdb_entry',       # PDB accession number
          'prior',           # compound found prior to Moonshot, eg. Covid 1
          'category',        # consortium vs. community
          'initial_screen',  # library screen
          'is_covalent',     # Is predicted covalent
          'purchasable',     # predicted purchasable
          'inspired_by',     # stated parent compounds
         ]].reset_index(drop=True).to_csv('moonshot-subset-core.csv')

moonshot[['Id', 
          'Ordered',
         'Made',
         'Assayed',
         'Maker',
         'order_date',
          'shipment_date',]].reset_index(drop=True).to_csv('moonshot-subset-logistics.csv')

moonshot[['Id', 
          'creator',         # human designer
          'rationale',       # their reasoning
          'Submission Notes',# //
          'N_chars',         # characters used
          'N_words',         # words used
          'classified_method', # predicted method
          'flesch',          # liguistic complexity
          'dale_chall',      # //
         ]].reset_index(drop=True).to_csv('moonshot-subset-words.csv')

moonshot[['Id', 
         'Covalent Fragment',
         'covalent_warhead',
         'Acrylamide',
         'Acrylamide Adduct',
         'Chloroacetamide',
         'Chloroacetamide Adduct',
         'Vinylsulfonamide',
         'Vinylsulfonamide Adduct',
         'Nitrile',
         'Nitrile Adduct',]
         ].reset_index(drop=True).to_csv('moonshot-subset-warhead_prediction.csv')

moonshot[['Id',
         'postera_SAScore',
         'postera_minNumSteps'
        ]].reset_index(drop=True).to_csv('moonshot-subset-SAScore.csv')
```
