This notebook demonstrates how to generate 3D coordinates for a molecule using RDKit and pass this to ODDT for computing 3D descriptors.

```python
from oddt.toolkits import rdk

from rdkit import Chem
from rdkit.Chem import AllChem

smiles = 'N#C/C=C/c1cc(Cl)cc(Oc2ccc(Cl)cc2OCCn2ccc(=O)[nH]c2=O)c1'
```

First we construct and RDKit Molecule and add explicit hydrogens - this is necessary for good conformer generation and some 3D descriptors (e.g. electroshape).

```python
mol = Chem.MolFromSmiles(smiles)
mol = Chem.AddHs(mol)
mol
```

Next, we generate a set of 3D coordinates for the molecule.

```python
AllChem.EmbedMolecule(mol, randomSeed=42)
mol
```

Finally, we construct an ODDT Molecule using our RDKit molecule.

```python
mol = rdk.Molecule(mol)
mol
```

We can verify that the ODDT Molecule has 3D coordinates.

```python
print(*mol.atoms, sep='\n')
```
